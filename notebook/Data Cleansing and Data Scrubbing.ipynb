```python
# Define a threshold for the maximum number of missing values
threshold = 250000

# Get the columns with missing values counts
missing_values = df.isnull().sum()

# Find columns with more than the specified threshold
columns_to_drop = missing_values[missing_values > threshold].index

# Drop the columns from the dataset
df = df.drop(columns=columns_to_drop)

# Now, data will contain only columns with fewer than 250,000 missing values.
```

```python
# Filter and get the list of numerical columns
numerical_columns = df.select_dtypes(include=[float, int]).columns.tolist()

# Replace missing values in numerical columns with their mean
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Replace missing values in categorical columns with their mode
categorical_columns = df.select_dtypes(include=[object]).columns.tolist()
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
for column in df.columns:
    unique_values_count = df[column].nunique()
    print(f"Column: {column}, Unique Values Count: {unique_values_count}")
```

```python
# Handling outliers

from scipy.stats import zscore

# Function to remove outliers using z-scores
def remove_outliers_zscore(data, column, threshold=3):
    z_scores = zscore(data[column])
    # Keep only rows where the absolute z-score is below the threshold
    data_before = data.shape[0]
    data = data[(abs(z_scores) < threshold) | (data[column].isna())]
    data_after = data.shape[0]

    print(f"Column: {column}, Rows before outliers removal: {data_before}, Rows after outliers removal: {data_after}")

    return data

# Iterate through numeric columns (int and float)
for column in df.select_dtypes(include=['int', 'float']).columns:
    df = remove_outliers_zscore(df, column)

# Reset the index
df = df.reset_index(drop=True)

```

```python
df
```

```python
# Handling categorical variables
non_numeric_columns = df.select_dtypes(exclude=['int', 'float']).columns

for column in non_numeric_columns:
    unique_values = df[column].unique()
    print(f"'{column}': {unique_values}")
```

```python

# Iterate through columns with datatype 'object'
for column in df.select_dtypes(include='object').columns:
    # Count the occurrences of each unique value in the column
    value_counts = df[column].value_counts()

    # Get the unique values that occur less than 20 times
    values_to_drop = value_counts[value_counts < 20].index

    # Drop the rows where the column value is in values_to_drop
    df = df[~df[column].isin(values_to_drop)]

# Reset the index after dropping rows
df = df.reset_index(drop=True)

```

```python
df
```

```python

# Function to check if a value is numeric
def is_numeric(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

# Convert object-type columns to int or float based on numeric content in at least one row
for column in df.select_dtypes(include='object').columns:
    # Check if any row in the column has purely numeric values
    if any(df[column].apply(is_numeric)):
        # Convert to float
        df[column] = pd.to_numeric(df[column], errors='coerce')

        # Convert to int if there are no decimal values, otherwise to float
        df[column] = df[column].astype(int) if all(pd.notna(value) and value.is_integer() for value in df[column]) else df[column].astype(float)


```

```python
# datatype of all the columns
for column in df.columns:
    print(f"{column}: {df[column].dtypes}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Extract features (excluding the target and labels)
features = df_filtered.drop(['Cluster_Labels', 'Risk_Level'], axis=1)

# Initialize an empty dictionary to store silhouette scores for each feature
silhouette_scores = {}

# Iterate over each feature and calculate silhouette score
for feature in features.columns:
    # Create a temporary DataFrame with the current feature and labels
    temp_df = pd.DataFrame(data={'Feature': features[feature], 'Cluster_Labels': df_filtered['Cluster_Labels']})

    # Initialize KMeans clustering
    kmeans = KMeans(n_clusters=len(df_filtered['Cluster_Labels'].unique()), random_state=42)

    # Fit KMeans on the temporary DataFrame
    temp_df['Cluster_Labels'] = kmeans.fit_predict(temp_df[['Feature']])

    # Calculate silhouette score for the current feature
    silhouette = silhouette_score(temp_df[['Feature']], temp_df['Cluster_Labels'])

    # Store the silhouette score in the dictionary
    silhouette_scores[feature] = silhouette

# Sort the features based on silhouette scores in descending order
sorted_features = sorted(silhouette_scores.items(), key=lambda x: x[1], reverse=True)

# Display the top features and their silhouette scores
for feature, score in sorted_features[:10]:
    print(f'{feature}: {score}')
```
